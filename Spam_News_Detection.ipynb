# Importing Libraries

```python
import pandas as pd
import numpy as np
import nltk
import re
from  nltk.corpus import stopwords
from  nltk.stem import WordNetLemmatizer
import string
```

```python
df_fake = pd.read_csv("Fake.csv")
df_true = pd.read_csv("True.csv")
```

```python
df_fake["label"] = 0
df_true["label"] = 1
```

```python
df_fake
```

```python
dataset1=df_fake[['text','label']]
dataset2=df_true[['text','label']]
```

```python
dataset1
```

```python
dataset2
```

```python
dataset=pd.concat([dataset1,dataset2])
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset['label'].value_counts()
```

```python
dataset=dataset.sample(frac = 1)
```

```python
dataset
```

```python
ps=WordNetLemmatizer()
```

```python
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
stopwords=stopwords.words('english')
```

```python
def clean_row(row):
    row=row.lower()
    row = re.sub(' [^a-zA-Z]', ' ', row)
    token = row.split()
    news = [ps.lemmatize(word) for word in token if not word in stopwords]
    cleanned_news = ' '.join(news)
    return cleanned_news
```

```python
dataset['text']
```

```python
dataset['text'] = dataset['text']. apply (lambda x : clean_row(x))
```

```python
dataset['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(max_features=1000,lowercase=False,ngram_range=(1,2))
```

```python
x=dataset.iloc[:35000,0]
y=dataset.iloc[:35000,1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data,train_label, test_label = train_test_split(x, y, test_size=0.2,random_state=0)
```

```python
vec_train_data = vectorizer.fit_transform(train_data)
```

```python
vec_train_data = vec_train_data.astype('float32').toarray()
```

```python
vec_test_data = vectorizer.fit_transform(test_data)
```

```python
vec_test_data = vec_test_data.toarray()
```

```python
vec_test_data.shape,vec_train_data.shape
```

```python
vec_train_data
```

```python
vec_test_data
```

```python
train_data = pd.DataFrame(vec_train_data, columns=vectorizer.get_feature_names_out())
testing_data = pd.DataFrame(vec_test_data, columns=vectorizer.get_feature_names_out())
```

```python
train_data
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf=MultinomialNB()
```

```python
clf.fit(train_data,train_label)
```

```python
y_pred=clf.predict(testing_data)
```

```python
test_label
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_label,y_pred)
```
